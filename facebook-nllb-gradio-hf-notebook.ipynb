```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import gradio as gr
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
translation_pipeline = pipeline('translation', 
                                model=model, 
                                tokenizer=tokenizer, 
                                src_lang="hin_Deva", 
                                tgt_lang='eng_Latn', 
                                max_length = 400)
```

```python
title = "NLLB Demo"
description = """
<p>
<center>
Hindi to English Translator using Facebook's NLLB
<img src="https://raw.githubusercontent.com/facebookresearch/fairseq/nllb/nllb.png" alt="rick"/>
</center>
</p>
"""
```

```python
def translate_gradio(input):
    result = translation_pipeline(input)
    return result[0]['translation_text']
```

```python
translate_interface = gr.Interface(fn = translate_gradio,
                                   inputs="text",
                                   outputs="text",
                                   title = title,
                                   description = description)
```

```python
translate_interface.launch()
```
